```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from seaborn import load_dataset
import pandas as pd
```

```python
ohdf = pd.read_csv("Datasets/healthcare-dataset-stroke-data.csv")
```

```python
ohdf.head()
```

```python
oh2df = ohdf[['gender', 'smoking_status','ever_married']]
oh2df = oh2df.dropna()
```

```python
oh2df.head()
```

```python
transformer = make_column_transformer(
    (OneHotEncoder(), ['gender', 'smoking_status', 'ever_married']),
    remainder='passthrough')
```

```python
transformed = transformer.fit_transform(oh2df)
transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names())
print(transformed_df.head())
```
