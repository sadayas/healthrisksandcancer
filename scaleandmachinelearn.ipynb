```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from seaborn import load_dataset
import pandas as pd
```

```python
##Get Dummies DF

ohe = pd.get_dummies(data=ohdf, columns=['gender', 'ever_married', 'work_type',
              'Residence_type', 'smoking_status'])
ohe.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
data_scaler = StandardScaler()
```

```python
stroke_data_scaled = data_scaler.fit_transform(ohe)
```

```python
import numpy as np
print(np.mean(stroke_data_scaled[:,0]))
print(np.std(stroke_data_scaled[:,0]))
```

```python
#Stroke Feature
oheX = ohe.copy()
oheX = oheX.drop("stroke", axis=1)
oheX.head()
```

```python
ohey = ohe["stroke"].values
ohey[:5]
```

```python
oheX_train, oheX_test, ohey_train, ohey_test = train_test_split(oheX, ohey, random_state=78)
```

```python
# Determine the shape of our training and testing sets.
print(oheX_train.shape)
print(oheX_test.shape)
print(ohey_train.shape)
print(ohey_test.shape)
```

```python
# Splitting into Train and Test sets into an 80/20 split.
oheX_train2, oheX_test2, ohey_train2, ohey_test2 = train_test_split(oheX, ohey, random_state=78, train_size=0.80)
```

```python
print(oheX_train2.shape)
print(oheX_test2.shape)
print(ohey_train2.shape)
print(ohey_test2.shape)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
oheX_scaler = scaler.fit(oheX_train)

# Scaling the data.
oheX_train_scaled = oheX_scaler.transform(oheX_train)
oheX_test_scaled = oheX_scaler.transform(oheX_test)
```

```python
# Create a random forest classifier.

rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(oheX_train_scaled, ohey_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(oheX_test_scaled)
```

```python
predictions
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(ohey_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(ohey_test, predictions)
acc_score
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(ohey_test, predictions))
```
